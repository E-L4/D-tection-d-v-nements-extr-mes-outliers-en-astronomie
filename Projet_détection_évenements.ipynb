```python
fluxdelétoile=np.nansum(data['FLUX'][:,y_center-1:y_center+2, x_center-1:x_center+2], axis=(1,2))
mask= (fluxdelétoile>np.nanmedian(fluxdelétoile)-5*np.nanstd(fluxdelétoile)) & (fluxdelétoile<np.nanmedian(fluxdelétoile)+5*np.nanstd(fluxdelétoile))

plt.plot(temps[mask],fluxdelétoile[mask])
plt.xlabel('temps en seconde')
plt.ylabel('Flux de l étoile')
plt.show()
```

```python
Fluxétoilesansnan=fluxdelétoile[(~np.isnan(fluxdelétoile))& (fluxdelétoile>0)]
medianflux=np.nanmedian(Fluxétoilesansnan)
stdflux=np.nanstd(Fluxétoilesansnan)

mask= (fluxdelétoile>medianflux-5*stdflux) & (fluxdelétoile<medianflux+5*stdflux)

plt.plot(temps[mask],fluxdelétoile[mask])
plt.xlabel('temps en seconde')
plt.ylabel('Flux de l\'étoile')
plt.show()
```

```python
mask= (fluxdelétoile>medianflux-3*stdflux) & (fluxdelétoile<medianflux+3*stdflux)
from scipy.ndimage import median_filter
fluxétoilelisse=median_filter(fluxdelétoile, size=5)
plt.plot(temps[mask],fluxétoilelisse[mask])
plt.xlabel('temps en seconde')
plt.ylabel('Flux de l\'étoile')
plt.show()
```
